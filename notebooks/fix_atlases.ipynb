```python
import nibabel as nib
import numpy as np
import os
```

```python
root = "/home/ali/Workspace/EpLink/Eplink"
atlases = ['Glasser_2016', 'Desikan', 'Schaefer2018_Kong2022_400', 'Yan2023_Kong2022_400']
atlases_path = os.path.join(root,"ISC-pipeline","resources","atlases_fsLR_32K")
```

```python
hemi = 'L'
atlas_file = os.path.join(atlases_path, f'{atlas}.32k.{hemi}.label.gii')
atlas_gii = nib.load(atlas_file)
atlas_labeltable = atlas_gii.labeltable.get_labels_as_dict()
atlas_data = atlas_gii.darrays[0].data
atlas_unique_data = np.unique(atlas_data)
atlas_labels = np.array(list(atlas_labeltable.keys()))

print(atlas_labeltable)
print(atlas_labels)
print(atlas_unique_data)
print(atlas_labels == atlas_unique_data)

# Data has no label 4 
# The only necessary change is to add a L_ prefix to the beginning of the values

for label in atlas_gii.labeltable.labels:
    label.label = f'{hemi}_{label.label}'

atlas_gii.labeltable.print_summary()
nib.save(atlas_gii, os.path.join(atlases_path, 'cleaned_versions', f'{atlas}.32k.{hemi}.label.gii'))

# Check the changes

atlas_gii = nib.load(os.path.join(atlases_path, 'cleaned_versions', f'{atlas}.32k.{hemi}.label.gii'))
atlas_labeltable = atlas_gii.labeltable.get_labels_as_dict()
atlas_data = atlas_gii.darrays[0].data
atlas_unique_data = np.unique(atlas_data)
atlas_labels = np.array(list(atlas_labeltable.keys()))


print(atlas_labeltable)
print(atlas_labels)
print(atlas_unique_data)
print(atlas_labels == atlas_unique_data)
```

# Schaefer 2018

```python
atlas = atlases[2]
```

```python
hemi = 'L'
atlas_file = os.path.join(atlases_path, f'{atlas}.32k.{hemi}.label.gii')
atlas_gii = nib.load(atlas_file)
atlas_labeltable = atlas_gii.labeltable.get_labels_as_dict()
atlas_data = atlas_gii.darrays[0].data
atlas_unique_data = np.unique(atlas_data)
atlas_labels = np.array(list(atlas_labeltable.keys()))

print(atlas_labeltable)
print(atlas_labels)
print(atlas_unique_data)
print(atlas_labels == atlas_unique_data)

# It has labels for both hemispheres also names are too long
# Will change the labeltable

atlas_gii.labeltable.labels = atlas_gii.labeltable.labels[:201]

for label in atlas_gii.labeltable.labels[1:]:
    label.label = f'{hemi}_{label.label[14:]}'

atlas_gii.labeltable.print_summary()
nib.save(atlas_gii, os.path.join(atlases_path, 'cleaned_versions', f'{atlas}.32k.{hemi}.label.gii'))

# Check the changes

atlas_gii = nib.load(os.path.join(atlases_path, 'cleaned_versions', f'{atlas}.32k.{hemi}.label.gii'))
atlas_labeltable = atlas_gii.labeltable.get_labels_as_dict()
atlas_data = atlas_gii.darrays[0].data
atlas_unique_data = np.unique(atlas_data)
atlas_labels = np.array(list(atlas_labeltable.keys()))


print(atlas_labeltable)
print(atlas_labels)
print(atlas_unique_data)
print(atlas_labels == atlas_unique_data)
```

```python
hemi = 'R'
atlas_file = os.path.join(atlases_path, f'{atlas}.32k.{hemi}.label.gii')
atlas_gii = nib.load(atlas_file)
atlas_labeltable = atlas_gii.labeltable.get_labels_as_dict()
atlas_data = atlas_gii.darrays[0].data
atlas_unique_data = np.unique(atlas_data)
atlas_labels = np.array(list(atlas_labeltable.keys()))

print(atlas_labeltable)
print(atlas_labels)
print(atlas_unique_data)
print(atlas_labels == atlas_unique_data)

# It has labels for both hemispheres also names are too long
# Will change the labeltable

atlas_gii.labeltable.labels = [atlas_gii.labeltable.labels[0]]+atlas_gii.labeltable.labels[201:]


for label in atlas_gii.labeltable.labels[1:]:
    label.label = f'{hemi}_{label.label[14:]}'

atlas_gii.labeltable.print_summary()
nib.save(atlas_gii, os.path.join(atlases_path, 'cleaned_versions', f'{atlas}.32k.{hemi}.label.gii'))

# Check the changes

atlas_gii = nib.load(os.path.join(atlases_path, 'cleaned_versions', f'{atlas}.32k.{hemi}.label.gii'))
atlas_labeltable = atlas_gii.labeltable.get_labels_as_dict()
atlas_data = atlas_gii.darrays[0].data
atlas_unique_data = np.unique(atlas_data)
atlas_labels = np.array(list(atlas_labeltable.keys()))


print(atlas_labeltable)
print(atlas_labels)
print(atlas_unique_data)
print(atlas_labels == atlas_unique_data)
```
